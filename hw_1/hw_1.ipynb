```python
import pandas
data = pandas.read_csv('titanic.csv')
```

```python
#get count of passengers on Titanic second way
count = data['PassengerId'].max()
print('count of passengers: ', count)
```

```python
#get percent of survival passangers
survival_count = len([d for d in data['Survived'] if d > 0])
print('count of survival passengers: ', survival_count)
survival_percent = round((survival_count * 100 / count), 2)
print('percent of survival passengers: ', survival_percent)
```

```python
#get percent of first class passengers
first_class = len(data[data['Pclass'] == 1])
print('count of first class passengers: ', first_class)
first_class_percent = round((first_class * 100 / count), 2)
print('percent of first class passengers: ', first_class_percent)
```

```python
#get age of passengers
median_age = data['Age'].median()
print('median age of passengers: ', median_age)
average_age = round(data['Age'].mean(), 2)
print('average age of passengers: ', average_age)
```

```python
#plot age of passengers
plt.rcParams['figure.figsize'] = [20, 8]
plt.title('Возраст пассажиров')
plt.xlabel('Номер пассажира')
plt.ylabel('Возраст пассажира')
plt.plot(data['Age'])
```

```python
#Get most popular female name
females = [name for name in data['Name'] if 'Mrs.' in name or 'Miss.' in name]
female_names = [name.split(', ')[1] for name in females]
pure_female_names = []
for name in female_names:
    name = name.replace('Miss. ', '')
    if '(' in name:
        name = name.split('(')[1][:-1]
    name = name.split(' ')[0]
    pure_female_names.append(name)
        
print(pure_female_names)
```

```python
c = collections.Counter()
for name in pure_female_names:
    c[name] += 1
    
most_popular_female_name = c.most_common()
print('most popular female name on Titanic: ', most_popular_female_name[0][0])
```
